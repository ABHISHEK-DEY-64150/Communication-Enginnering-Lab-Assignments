**Step1: Sampling**\
Let ,\
 n be the number of samples per interval.\
fs be the sampling frequency.


```python
def create_sample_signal(t,A,f):

  y = A*np.sin( 2 * np.pi * f * t )

  return y
```

```python
fs = 25

steps = 1.0/fs 

n = np.arange(0,2.0,steps)

ys = create_sample_signal(n, 6, 2)

plt.title(' The sampled analogue signal at fs = 15', fontsize = 15)
plt.stem(n,ys)
plt.show()
```

**Step2: Quantization**

we have to perfrom the following :

1.we have to partition the signal value with height delta of L number of levels.between Max amplitude and Min amplitude.

2.Decrease the partition value by delta/2 .(codebook)

3.Assign every codebook interval the corresponding partition number or the quantized values.

lets,consider the number of bits per level = 3 ,so the total number of levels would be pow(2,3)= 8.


```python
n_bits = 3
L = 2 ** n_bits

A = 6

mx = A
mn = -A 

delta  = (mx-mn)/L 

prtion = np.arange(mn,mx,delta)

cdbook = np.arange(mn-(delta/2),mx+(delta/2),delta)

yq = np.copy(ys)

encode = np.copy(ys)

for i in range(len(prtion)):
  l = cdbook[i]
  r = cdbook[i + 1]
  for j in range(len(ys)):
   
    if(i != len(prtion) - 1 and l <= ys[j] and ys[j] < r):
      yq[j] = prtion[i]
      encode[j] = i
   
    if(i == len(prtion) - 1 and l <= ys[j]):
      yq[j] = prtion[i]
      encode[j] = i


for i in range(len(yq)):
    yq[i] = yq[i] + (delta / 2) 

plt.title(' The sampled analogue signal at fs = 15', fontsize = 15)
plt.stem(n,ys)
plt.show()


plt.title(' The sampled  & quantized analogue signal', fontsize = 15)
plt.stem(n,yq)
plt.show()
```

**Step3: Encoding**

we have to perform the following

1.Convert the assigned decimal codec to the binary codec.\
2.Append all the binary codec into an array of 0 and 1.\
2.plot a step graph



```python
def decimalToBinary(val, bit):
    binary = "{0:b}".format(int(val))
    if(len(binary) > bit):
      return binary
    extra_zero = bit - len(binary)
    return extra_zero * "0" + binary

    
```

```python
binary_codec = []
for i in range(len(encode)):
  binary = decimalToBinary(encode[i], n_bits)
  for j in range(len(binary)):
    binary_codec.append(binary[j])

pos = np.arange(len(binary_codec))


plt.figure(figsize=(20, 5))
plt.title('The digital signal', fontsize = 15)

plt.ylim(-1, 3)
plt.step(pos, binary_codec)

plt.show()
```

**Frequnecy Shift Keying (FSK)->**

```python
def uNRZ(bits, N, bit_rate, T, Time):
  
  x = np.array([])
  for i in range(N):
    temp = bits[i]* np.ones((T))
    x = np.append(x, temp)
  t = np.linspace(0,N*Time,len(x))
  return [t, x]
```

```python

bits = np.array([0,1,0,1,0,1,1,1,0,0,1,0,1,0,1])
Time = 10e-6   
bit_rate = 1/Time 
N = len(bits)   
T = 100
t,x = uNRZ(bits, N, bit_rate, T, Time)

plt.figure(figsize=(20, 5))
plt.plot(t,x)
plt.show()
```

```python
# Plot an image of the spectrogram y, with the axis labeled with time tl,
# and frequency fl
#
# t_range -- time axis label, nt samples
# f_range -- frequency axis label, nf samples
# y -- spectrogram, nf by nt array
# dbf -- Dynamic range of the spect

def sg_plot( t_range, f_range, y, dbf = 60) :
    eps = 1e-3
    
    # find maximum
    y_max = abs(y).max()
    
    # compute 20*log magnitude, scaled to the max
    y_log = 20.0 * np.log10( abs( y ) / y_max + eps )
    
    fig=figure(figsize=(15,6))
    
    plt.imshow( np.flipud( 64.0*(y_log + dbf)/dbf ), extent= t_range  + f_range ,cmap=plt.cm.gray, aspect='auto')
    plt.xlabel('Time, s')
    plt.ylabel('Frequency, Hz')
    plt.tight_layout()


def myspectrogram_hann_ovlp(x, m, fs, fc,dbf = 60):
    # Plot the spectrogram of x.
    # First take the original signal x and split it into blocks of length m
    # This corresponds to using a rectangular window %
    
    
    isreal_bool = isreal(x).all()
    
    # pad x up to a multiple of m 
    lx = len(x);
    nt = (lx + m - 1) // m
    x = append(x,zeros(-lx+nt*m))
    x = x.reshape((m//2,nt*2), order='F')
    x = concatenate((x,x),axis=0)
    x = x.reshape((m*nt*2,1),order='F')
    x = x[r_[m//2:len(x),ones(m//2)*(len(x)-1)].astype(int)].reshape((m,nt*2),order='F')
    
    
    xmw = x * hanning(m)[:,None];
    
    
    # frequency index
    t_range = [0.0, lx / fs]
    
    if isreal_bool:
        f_range = [ fc, fs / 2.0 + fc]
        xmf = np.fft.fft(xmw,len(xmw),axis=0)
        sg_plot(t_range, f_range, xmf[0:m//2,:],dbf=dbf)
        print (1)
    else:
        f_range = [-fs / 2.0 + fc, fs / 2.0 + fc]
        xmf = np.fft.fftshift( np.fft.fft( xmw ,len(xmw),axis=0), axes=0 )
        sg_plot(t_range, f_range, xmf,dbf = dbf)
    
    return t_range, f_range, xmf
```
